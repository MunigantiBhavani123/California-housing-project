##10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
housing.loc[housing["total_bedrooms"] <=10, "total_bedroom_size"] = "small"  #I used loc method it will traces based on clm name
housing.loc[housing["total_bedrooms"] <=1000, "total_bedroom_size"] = "medium"
housing.loc[housing["total_bedrooms"] >1000, "total_bedroom_size"] = "large"
housing
```
